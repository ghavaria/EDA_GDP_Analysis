```python
#indexinig and extracting data for (% Growth over previous year)
GDP_df.set_index(['Items  Description','Duration'], inplace = True)
GDP_df.loc['(% Growth over previous year)']
```

```python
# resetting the index
GDP_df_temp = GDP_df.loc['(% Growth over previous year)']
GDP_df_temp.reset_index(inplace = True)
GDP_df_temp
```

```python
statewise_growth = {} #creating empty dictionary

for state in GDP_df_temp.columns:
    if state not in ['Duration']:
        tmp = GDP_df_temp[['Duration', state]]
        tmp.dropna(inplace = True)
        tmp.reset_index(inplace= True)
        plt.title("% Growth of "+state)
        plt.xlabel("year")
        plt.ylabel("% Growth over previous year")
        plt.ylim((-10, 30))
        
        plt.scatter(x=tmp['Duration'], y=tmp[state], c = 'red')
        print(np.polyfit(tmp.index,tmp[state],1))
        m ,c = np.polyfit(tmp.index,tmp[state],1)
        
        statewise_growth[state]=round(m,3)
        
        plt.plot(tmp.index, m*tmp.index+c, markersize = 1)
        plt.show()
statewise_growth
```

# GDP_Assignment_Part I A 2nd sub question

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

GDP_df = pd.read_csv("C:\GDP_Assignment\Data_1.csv")

# Dropping West bengal since it has all NA values
GDP_df = GDP_df.drop('West Bengal1', axis=1)

#Dropping Union Territories and All India data 
GDP_df = GDP_df.drop(['Delhi','Chandigarh','Andaman & Nicobar Islands','Puducherry', 'All_India GDP'], axis=1)

#GDP_df.head()
```

```python
# Indexing and f;inding data for GSDP - CURRENT PRICES (` in Crore)
GDP_df.set_index(['Items  Description','Duration'], inplace = True)
GDP_df.loc['GSDP - CURRENT PRICES (` in Crore)']

```

```python
#resetting Indexing 
GDP_df_temp = GDP_df.loc['GSDP - CURRENT PRICES (` in Crore)']
GDP_df_temp.reset_index(inplace = True)
GDP_df_temp
```

```python
#extracting values for Duration 2015-16 only 
df= GDP_df_temp[4:5]
df.head()
```

```python
#Transposing the rows into columns
GDP = df.T
GDP
```

```python
#resetting index on transposed datafreame 
GDP.reset_index(inplace = True)
GDP
```

```python
# updating the header with 0th row 
GDP.columns = GDP.iloc[0]
GDP
```

```python
#dropping 0th row
GSDP = GDP.drop(GDP.index[0])
GSDP
```

```python
#renaming the columns
GSDP.rename(columns = {'Duration':'States', '2015-16':'GSDP'}, inplace = True) 
GSDP
```

```python
# updating the GSDP coulmn in decending order
final_GDP = GSDP.sort_values('GSDP',ascending=False)
final_GDP
```

```python
#Dropping NaN values as it is not required for our Analysis
GDP_State = final_GDP.dropna()
GDP_State
```

```python
# Plotting bar graph which is easier to read and compare the data 
plt.figure(figsize=(10,10))
sns.barplot(x='GSDP', y='States', data=GDP_State)
plt.title("GDP of the states for the year 2015-16")
plt.show()
```

```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'C:\GDP_Assignment\Data I B'
#path = r'C:\GDP_Assignment\dump'
```

```python
all_files = glob.glob(path+'/*.csv')
all_files
```

```python
List = []
for location in all_files:
    df = pd.read_csv(location, index_col = None, header=0, encoding = "cp1252")
    df = df[['Item', '2014-15']]
    df = df.set_index('Item')
    df = df.T
    df.loc[:,'States'] = location.split("\\")[-1].split(".csv")[0].split("-GSVA_cur")[0].split("NAD-")[-1]
    List.append(df)
```

```python
List
```

```python
GDP = pd.concat(List, axis =0, ignore_index = True)
GDP
```

```python
GSDP = GDP[['States', 'Per Capita GSDP (Rs.)']] 
GSDP
```

```python
Data = GSDP.sort_values('Per Capita GSDP (Rs.)',ascending=False)
Data
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(x='Per Capita GSDP (Rs.)', y='States', data = Data)
plt.title("GDP Per capita of all States")
plt.show()
```
